# Project Description
The House Construction Cost Calculator System is a Python-based application designed to estimate the total cost of building a house.
It allows users to select a house type such as Independent, G+1, or G+2 and enter the total area in square feet.
Based on the selected type, the system assigns a predefined cost per square foot.
The total construction cost is calculated using the formula: Area × Cost per sq.ft.
The project is developed using Object-Oriented Programming concepts like classes, constructors, and methods.
It helps reduce manual calculation errors and provides a simple, efficient way to estimate construction expenses.

```python
class House:
    def __init__(self, owner_name, house_type, area):
        self.owner_name = owner_name
        self.house_type = house_type
        self.area = area

    def calculate_cost(self):
        if self.house_type == "1":
            cost_per_sqft = 35000   
            type_name = "INDEPENTANT HOSUE"
        elif self.house_type == "2":
            cost_per_sqft = 57000 
            type_name = "G+1 House"
        elif self.house_type == "3":
            cost_per_sqft = 70000  
            type_name = "G+2 House"
        else:
            print("No House Type!")
            return None,None
            

        total_cost = self.area * cost_per_sqft
        return total_cost, type_name

    def display_details(self):
        total_cost, type_name = self.calculate_cost()

        if total_cost is not None:
            print("\n===== House Construction Details =====")
            print("Owner Name:", self.owner_name)
            print("House Type:", type_name)
            print("Area (sq.ft):", self.area)
            print("Total Construction Cost: ", total_cost)


def main():
    while True:
        print("\n===== House Construction System =====")
        print("1. Independent House(35000 per sq.ft)")
        print("2. G+1 House (57000 per sq.ft)")
        print("3. G+2 House (70000 per sq.ft)")
        

        choice = input("Select House Type: ")

        if choice >= "4":
            print("Did not build the house!")
            break

        
        owner = input("Enter Owner Name: ")
        area = float(input("Enter Area in sq.ft: "))

        house = House(owner, choice, area)
        house.display_details()
main()
```
